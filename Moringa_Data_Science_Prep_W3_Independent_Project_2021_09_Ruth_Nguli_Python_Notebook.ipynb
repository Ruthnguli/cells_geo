```python

# we change the column Villes which is city as per the discription to display a new name City for farmiliarity
df_new = df.rename(columns={'VILLES': 'CITY', 'SITE_CODE': 'SITE_ID'})
df_new
```

```python
#loading data our CDR discription dataset for TELECOME DATASETS
df = pd.read_excel('CDR_description.xlsx')
df
```

```python

#loading and creating Dataframe from Telecome dataset1 url=https://raw.githubusercontent.com/Ruthnguli/cells_geo/main/Telcom_dataset.csv

url = "https://raw.githubusercontent.com/Ruthnguli/cells_geo/main/Telcom_dataset.csv"
df_1 = pd.read_csv(url)
#viewing the first 10 rows
df_1.head(100)
```

```python

#Update column PRODUTc and DATATIME to PRODUCT and DATA_TIME as per the description

df_1_new = (df_1.rename(columns={'PRODUTC': 'PRODUCT', 'DATETIME': 'DATE_TIME'}))
df_1_new.head(100)
```

```python
# #loading and creating Dataframe from Telecome dataset2 url =https://raw.githubusercontent.com/Ruthnguli/cells_geo/main/Telcom_dataset2.csv
url = "https://raw.githubusercontent.com/Ruthnguli/cells_geo/main/Telcom_dataset2.csv"
df_2 = pd.read_csv(url)
#viewing the first 10 rows
df_2.head(10)
```

```python
#updating columns DW_A_NUMBER and 	DW_B_NUMBER	to DW_A_NUMBER_INT	DW_B_NUMBER_INT to march the description
df_2_new = (df_2.rename(columns={'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER': 'DW_B_NUMBER_INT'}))
df_2_new
```

```python

# loading and creating Dataframe from Telecome dataset3 url = https://raw.githubusercontent.com/Ruthnguli/cells_geo/main/Telcom_dataset3.csv

url = "https://raw.githubusercontent.com/Ruthnguli/cells_geo/main/Telcom_dataset3.csv"

df_3 = pd.read_csv(url)
#viewing the first 10 rows
df_3.head(10)
```

```python
#updating columns CELLID and SIET_ID to CELL_ID and SITE_ID to march the description
df_3_new = (df_3.rename(columns={'CELLID': 'CELL_ID', 'SIET_ID': 'SITE_ID'}))
df_3_new
```

```python
#merge the 3 CDR TELECOM DATAFRAMES

CDR_TELECOM = pd.concat([df_1_new, df_2_new,df_3_new], axis=0)
CDR_TELECOM
```

```python
#Merging the new marged Dataset CDR DATAFRAME with the cell geo dataframe


dfm = CDR_TELECOM.merge(df_new, how='left', on='SITE_ID')
dfm

```

```python
# Data cleaning
#Identify and drop duplicate rows

dfdup = dfm.duplicated()
dfdup
```

```python
#Droping the duplicated rows

dfnew = dfm.drop_duplicates()
dfnew
```
